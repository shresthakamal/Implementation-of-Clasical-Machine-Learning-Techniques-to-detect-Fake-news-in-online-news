```python
def Punctuation(content):
    cnt_in_descr=0
    #a=len(description)
    check_descr=[]                    
    i=0
    length=len(data[content])
    col=data[content]
    for i in range(length):
        #Loop from index 0 to last index
        check_descr=word_tokenize(col[i])     #Tokenization in row[i] in every step
        for w in check_descr:                         #If the Array of words(word tokens) of row[i]
            if w in punctuation:                      #contains punctuation, increase counter    
                cnt_in_descr+=1
        print("Punctuation in ",content,"row ", i, " is: ",cnt_in_descr)
        cnt_in_descr=0
        
Punctuation('title')        #Takes variable as string
Punctuation('description')
```

```python
#Removing stop words from title of the news
def Stop_Removal(content):
    i=0
    filtered_content=[]
    col=data[content]
    length=len(col)
    
    for i in range(length):
        filtered_content=[]
        words_content= word_tokenize(col[i])
        for w in words_content:
            if w not in stop_words:
                filtered_content.append(w)

        words_content=filtered_content 
        print("Text for ",content, i,"=", words_content)        #Stop Words removed. To store in list and To print Uncomment 
                    #Uncomment line below for normal operation
        #print("The filtered", content,i,"is",filtered_content)  
        #words_title[i]=filtered_title    #"list assignment index out of range" ERROR while trying to store in rows
        #print(words_title)        #Stop Words removed. Out of range 

    #filtered_description= w for w in words if not w in stop_words    #Tokenzation

Stop_Removal('title')
Stop_Removal('description')
```

```python
def Sentiment(content):
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        blob=TextBlob(lst[i])
        print("The sentiment of ",content, i,"=",blob.sentiment)
    
Sentiment('title')
Sentiment('description')
```

```python
def Incorrect_Spaces_After_Comma(content):
    i=0
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        sent=lst[i]
        print("The incorrect spaces after comma in ",content, i,"=",len(re.findall(('\w*\,\s{2,}'),sent)))

Incorrect_Spaces_After_Comma('title')
Incorrect_Spaces_After_Comma('description')
```

```python
def Incorrect_Spaces_After_Full_Stop(content):
    i=0
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        sent=lst[i]
        print("The incorrect spaces after fullstop in ",content, i,"=",len(re.findall(('\w*\.\s{2,}'),sent)))

Incorrect_Spaces_After_Full_Stop('title')
Incorrect_Spaces_After_Full_Stop('description')
```

```python
def Incorrect_Fullstop_End_of_Title(content):
    i=0
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        sent=lst[i]
        print("Is full stop present at the end of title? For row",content, i, "=", len(re.findall('\w*\.$',sent)))

Incorrect_Fullstop_End_of_Title('title')
```

```python
def Incorrect_Use_of_Punctuation(content):
    i=0
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        sent=lst[i]
        print("The incorrect use of expressions in row",content, i, "=", len(re.findall('\[!"#[$%&<?>{}]_+=-/^.`*|()@\~:;]{2,}',sent)))

Incorrect_Use_of_Punctuation('title')
```
