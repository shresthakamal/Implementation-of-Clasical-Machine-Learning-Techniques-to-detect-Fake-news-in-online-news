```python
def Punctuation(content):
    cnt_in_descr=0
    #a=len(description)
    check_descr=[]                    
    i=0
    length=len(data[content])
    col=data[content]
    for i in range(length):
        #Loop from index 0 to last index
        check_descr=word_tokenize(col[i])     #Tokenization in row[i] in every step
        for w in check_descr:                         #If the Array of words(word tokens) of row[i]
            if w in punctuation:                      #contains punctuation, increase counter    
                cnt_in_descr+=1
        print("Punctuation in ",content,"row ", i, " is: ",cnt_in_descr)
        cnt_in_descr=0
        
Punctuation('title')        #Takes variable as string
Punctuation('description')
```

```python
#Removing stop words from title of the news
def Stop_Removal(content):
    i=0
    filtered_content=[]
    col=data[content]
    length=len(col)
    
    for i in range(length):
        filtered_content=[]
        words_content= word_tokenize(col[i])
        for w in words_content:
            if w not in stop_words:
                filtered_content.append(w)

        words_content=filtered_content 
        print("Text for ",content, i,"=", words_content)        #Stop Words removed. To store in list and To print Uncomment 
                    #Uncomment line below for normal operation
        #print("The filtered", content,i,"is",filtered_content)  
        #words_title[i]=filtered_title    #"list assignment index out of range" ERROR while trying to store in rows
        #print(words_title)        #Stop Words removed. Out of range 

    #filtered_description= w for w in words if not w in stop_words    #Tokenzation

Stop_Removal('title')
Stop_Removal('description')
```

```python
def Sentiment(content):
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        blob=TextBlob(lst[i])
        print("The sentiment of ",content, i,"=",blob.sentiment)
    
Sentiment('title')
Sentiment('description')
```

```python
import re
def Double_Space(content):
    lst=data[content]
    length=len(lst)
    i=0
    for i in range(length):
        data=lst[i]
        print("The double spaces in ",content, i, "=",len(re.findall('\s\s+',data)))
        print("The no of space after comma in ",content, i,"=",len(re.findall(('\w\,\w'),data)))
        print("The number of capitalized words of",content, i, "=",len(re.findall(('[A-Z][a-z]*'),data)))
        print("*"*25)
        
Double_Space('title')
Double_Space('description')
```
